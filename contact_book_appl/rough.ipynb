```python
import pandas as pd

def add_contact(name, phone_number, address, remarks, df):
    """Adds a new contact to the DataFrame.

    Args:
        name (str): The contact's name.
        phone_number (str): The contact's phone number.
        address (str): The contact's address.
        remarks (str): Any additional remarks about the contact.
        df (pd.DataFrame): The DataFrame to store the contact information.

    Returns:
        None
    """

    # Validate input
    if not isinstance(name, str) or not name:
        raise ValueError("Name must be a non-empty string.")
    if not isinstance(phone_number, str) or not phone_number:
        raise ValueError("Phone number must be a non-empty string.")
    if not isinstance(address, str) or not address:
        raise ValueError("Address must be a non-empty string.")
    if not isinstance(remarks, str):
        raise ValueError("Remarks must be a string.")

    # Check if contact already exists
    if name in df['name'].values:
        raise ValueError("Contact already exists.")
    if phone_number in df['phone_number'].values:
        raise ValueError("phone_number already exists.")
    

    # Create a new row for the contact
    new_contact = pd.Series({
        'name': name,
        'phone_number': phone_number,
        'address': address,
        'remarks': remarks
    })

    # Append the new row to the DataFrame
    df = pd.concat([df, new_contact.to_frame().T], ignore_index=True)

    print("Contact added successfully.")

    return df

## To find contact function__________________________________________________________________
def find_contact(name, df):
    """Finds a contact by name in the DataFrame."""
    contact = df[df['name'] == name]
    if contact.empty:
        print("Contact not found.")
    else:
        print(contact)

# Example Usage
# find_contact('John Doe', df)


## to Update contact function__________________________________________________________
def update_contact(name, new_phone_number, new_address, new_remarks, df):
    """Updates a contact in the DataFrame."""
    if name not in df['name'].values:
        print("Contact not found.")
    else:
        df.loc[df['name'] == name, ['phone_number', 'address', 'remarks']] = [new_phone_number, new_address, new_remarks]
        print("Contact updated successfully.")
    return df

# Example Usage
# df = update_contact('John Doe', '1234567890', 'New Address', 'Updated remarks', df)

## To delete contact function__________________________________________________________________
def delete_contact(name, df):
    """Deletes a contact from the DataFrame."""
    if name not in df['name'].values:
        print("Contact not found.")
    else:
        df = df[df['name'] != name]
        print("Contact deleted successfully.")
    return df

# Example Usage
# df = delete_contact('John Doe', df)






```
